```python
# ChatGPT temperature controls the randomness and creativity of responses, typically ranging from 0 to 1 (or up to 2 in some API settings). 
# A lower value (0.0) makes output more deterministic, focused, and consistent, while a higher value (1.0) increases creativity, randomness, 
# and risk of nonsensical output. 

prompt = "Write a short, creative story (up to two paragraphs) about a robot discovering a flower."

# Generate 4 versions: 
# Temp 0.0 (deterministic), 
# Temp 0.5 (balanced), 
# Temp 0.8 (creative), 
# and Temp 1.0 (highly imaginative/unpredictable)." 

print("=== Temperature Demo ===\n")
print("Low temperature (0.2) - Focused and consistent:")
response = (generate_text(prompt, temperature=0.2))
render_markdown(response)
```

```python
print("=== Temperature Demo ===\n")
print("\nHigh temperature (1) - Creative and varied:")
response = generate_text(prompt, temperature=1)
render_markdown(response)
```

```python
"""
Demonstrates how max_tokens controls response length.
"""

prompt = "Explain what machine learning is."

print("=== Max Tokens Demo ===\n")
print("Short response (50 tokens):")
response = generate_text(prompt, max_tokens=50)
render_markdown(response)
```

```python
print("\nLonger response (200 tokens):")
response = generate_text(prompt, max_tokens=200)
render_markdown(response)
```

```python
"""
Demonstrates structured JSON output.
"""
prompt = "Analyze the sentiment of this review: 'This product exceeded my expectations!'"
system_instruction = "You are a sentiment analyzer. Always respond with valid JSON containing 'sentiment' and 'confidence' fields."

print("=== JSON Mode Demo ===\n")
result = generate_text(
    prompt, 
    system_instruction=system_instruction,
    response_format={"type": "json_object"}
)
print("Structured JSON output:")
print(result)
print("\n")
```

```python
"""
Demonstrates streaming responses for real-time feedback.
"""
prompt = "Write a short poem about coding."

print("=== Streaming Demo ===\n")
print("Streaming response (appears word-by-word):")

stream = generate_text(prompt, stream=True)
for chunk in stream:
    print(chunk, end="", flush=True)
print("\n\n")
```

```python
"""
Demonstrates frequency penalty to reduce repetition.
"""

prompt = "Write a poem about nature where every word starts with Z"

# "List 5 benefits of exercise."

print("=== Frequency Penalty Demo ===\n")
print("Without penalty (may repeat words):")
response = generate_text(prompt, frequency_penalty=0.0)
render_markdown(response)
```

```python
print("\nWith penalty (more variety):")
response = generate_text(prompt, frequency_penalty=1.0)
render_markdown(response)
```

```python
"""
Demonstrates using stop sequences to control output.
"""
prompt = """Write a Python function to calculate fibonacci numbers.
After the function, write 5 test cases marked with # TEST CASES"""

# Stop before test cases - you just want the function
code = generate_text(prompt)
render_markdown(code)
```

```python
code = generate_text(prompt, stop="# TEST CASES")
render_markdown(code)
# Now you have clean function code without tests mixed in
```

```python
"""
Demonstrates reproducible outputs with seed.

Purpose:
Testing & Debugging - When building an app, you want consistent outputs so you can:

1. Test that your code handles responses correctly
2. Debug issues without the model giving different answers each time
3. Compare changes to your prompt (A/B testing)
"""
prompt = "Generate a random product name."
seed_value = 12345

print("=== Seed Demo (Reproducibility) ===\n")
print(f"First call with seed={seed_value}:")
print(generate_text(prompt, seed=seed_value, temperature=1.0))
print(f"\nSecond call with same seed={seed_value}:")
print(generate_text(prompt, seed=seed_value, temperature=1.0))
print(f"\nCall with different seed:")
print(generate_text(prompt, seed=67890, temperature=1.0))
print("\n")
```
