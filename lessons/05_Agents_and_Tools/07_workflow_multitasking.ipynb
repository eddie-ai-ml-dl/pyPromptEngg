## ğŸ“‹ Use Case: Multi-Platform Social Media Campaign Generation

Imagine you want to launch a promotional campaign for a new product. Instead of having one generalist AI agent create all the content sequentially, you can use the **parallelization pattern** to create specialized content for different platforms **simultaneously**.

### The Overall Task
Generate a cohesive social media campaign for a new product based on:
- Product description
- Target audience
- Key features

---

## ğŸ¤– Specialized Agents

This main task is broken down into parallel sub-tasks, each handled by a specialized AI agent:

### Agent 1: The Twitter Specialist
- **Expertise**: Crafting short, punchy, and engaging content suitable for Twitter/X
- **Output**: A series of tweets that are concise and include relevant hashtags
- **Character Limit**: Under 280 characters per tweet

### Agent 2: The LinkedIn Professional
- **Expertise**: Professional tone suitable for B2B platform
- **Output**: Detailed, business-oriented post highlighting value proposition
- **Focus**: Business benefits, innovation, professional networking

### Agent 3: The Blog Post Writer
- **Expertise**: Long-form content creation
- **Output**: Comprehensive blog post (500+ words)
- **Content**: Deep dive into features, benefits, and use cases

---

## âš™ï¸ The Workflow in Action

### 1. Input
A single brief is provided to all agents, containing:
- Product name
- Short description
- Key features
- Target audience

### 2. Parallel Execution
The three agents (Twitter, LinkedIn, and Blog) all start working **simultaneously**.
- âœ… No waiting for other agents to finish
- âœ… Tasks are independent
- âœ… Drastically reduces total execution time

### 3. Aggregation
The outputs from each agent are collected and presented together, ensuring:
- Consistency in messaging across platforms
- Complete campaign coverage
- Efficient use of time and resources

```python
# ============================================================================
# PARALLEL WORKFLOW ORCHESTRATION
# ============================================================================
# This cell demonstrates how to orchestrate multiple AI agents to run in
# parallel, significantly reducing total execution time compared to sequential
# processing.
# ============================================================================

import time
from domo_agents.twitter_agent import TwitterAgent
from domo_agents.linkedin_agent import LinkedInAgent
from domo_agents.blog_post_agent import BlogPostAgent
from workflows.parallel_workflow import ParallelWorkflow


def main():
    """
    Orchestrate the social media campaign generation using parallel execution.
    
    This function demonstrates the parallelization pattern where multiple
    specialized AI agents work simultaneously on independent tasks, all
    processing the same input brief but generating platform-specific content.
    
    Workflow Steps:
        1. Define input product brief
        2. Initialize specialized agents
        3. Execute all agents in parallel
        4. Aggregate and display results
    
    Returns:
        None: Results are printed to console
    """
    print("ğŸš€ Starting social media campaign generation...\n")
    print("=" * 70)

    # ========================================================================
    # STEP 1: Define the input for the agents
    # ========================================================================
    # All agents will receive the same product brief but will generate
    # content optimized for their specific platform/purpose
    
    product_brief = {
        "name": "Synergy AI",
        "description": "An AI-powered platform for seamless team collaboration.",
        "audience": "tech professionals and project managers",
        "features": [
            "Automated Task Assignment",
            "Predictive Analytics",
            "Real-time Reporting"
        ]
    }
    
    print("ğŸ“‹ Product Brief:")
    print(f"   Name: {product_brief['name']}")
    print(f"   Description: {product_brief['description']}")
    print(f"   Target Audience: {product_brief['audience']}")
    print(f"   Features: {', '.join(product_brief['features'])}")
    print("=" * 70 + "\n")

    # ========================================================================
    # STEP 2: Initialize the specialized agents
    # ========================================================================
    # Each agent is a specialized AI that knows how to create content
    # for a specific platform using OpenAI's API
    
    all_agents = [
        TwitterAgent(),      # Creates short, engaging tweets
        LinkedInAgent(),     # Creates professional B2B content
        BlogPostAgent()      # Creates long-form detailed articles
    ]
    
    print(f"âœ… Initialized {len(all_agents)} specialized agents:")
    for agent in all_agents:
        print(f"   â€¢ {agent.name}")
    print()

    # ========================================================================
    # STEP 3: Initialize and run the parallel workflow
    # ========================================================================
    # The ParallelWorkflow uses ThreadPoolExecutor to run all agents
    # simultaneously, allowing concurrent API calls to OpenAI
    
    workflow = ParallelWorkflow(agents=all_agents)
    
    print("â±ï¸  Starting parallel execution...")
    start_time = time.time()
    
    # The 'run' method will:
    # 1. Submit all agent tasks to a thread pool
    # 2. Execute them in parallel (all API calls happen at the same time)
    # 3. Block until all agents have completed
    # 4. Return aggregated results
    campaign_content = workflow.run(product_brief)
    
    end_time = time.time()
    execution_time = end_time - start_time
    
    print(f"\nâœ¨ All agents have finished in {execution_time:.2f} seconds")
    print("=" * 70)

    # ========================================================================
    # STEP 4: Display the aggregated results
    # ========================================================================
    # Results are organized by agent name, showing all generated content
    
    print("\nğŸ“Š GENERATED SOCIAL MEDIA CAMPAIGN CONTENT")
    print("=" * 70 + "\n")
    
    for agent_name, content in campaign_content.items():
        print(f"â”Œâ”€ Content by {agent_name} " + "â”€" * (70 - len(agent_name) - 15))
        print(f"â”‚")
        # Print content with proper indentation
        for line in content.split('\n'):
            print(f"â”‚ {line}")
        print(f"â””" + "â”€" * 69)
        print()
    
    print("=" * 70)
    print(f"âœ… Campaign generation complete! Total time: {execution_time:.2f}s")
    print("=" * 70)
```

```python
# ============================================================================
# EXECUTE THE PARALLEL WORKFLOW
# ============================================================================
# Run the main function to see all agents executing in parallel
# Watch the console output to observe concurrent execution
# ============================================================================

main()
```
