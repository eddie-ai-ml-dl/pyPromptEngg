```python
from langchain_community.document_loaders import WebBaseLoader
url = "https://www.ilri.org/about/people/governance"
# Load the document from the URL
loader = WebBaseLoader(url)
docs = loader.load()

# docs is now available for further processing
# Each document has .page_content and .metadata attributes
for doc in docs:
    print(doc.page_content)
    print(doc.metadata)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Initialize the chat model
chat = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# Create a prompt template for extracting names and titles
extraction_prompt = PromptTemplate.from_template(
    """Extract all people's names and their titles from the following text.
Format the output as a markdown table with headers: Name | Title

If no title is found, write "Not specified" in the Title column.

Text:
{content}

Extracted People (Markdown Table):"""
)

# Build the extraction chain
extraction_chain = extraction_prompt | chat | StrOutputParser()

# Process all loaded documents
all_results = []
for doc in docs:
    result = extraction_chain.invoke({"content": doc.page_content})
    all_results.append(result)
    print(f"\n=== Extracted from {doc.metadata.get('source', 'unknown')} ===")
    print(result)
```
