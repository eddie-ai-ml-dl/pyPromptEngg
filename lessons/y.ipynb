```python
from langchain_community.document_loaders import WebBaseLoader
url = "https://openai.com/index/hardening-atlas-against-prompt-injection/"
# Load the document from the URL
loader = WebBaseLoader(url)
docs = loader.load()

# docs is now available for further processing
# Each document has .page_content and .metadata attributes
for doc in docs:
    print(doc.page_content)
    print(doc.metadata)
```
